```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
json_file = "./arxiv_db.json"
arxiv_db = pd.read_json(json_file, lines=True)
arxiv_db
```

```python
arxiv_paper_content = arxiv_db[['authors','update_date','title','categories','abstract']]
arxiv_paper_content.head()
```

```python
#pd.set_option('display.max_rows', 50)
#arxiv_paper_content["categories"].unique()
astro_ph_papers = arxiv_paper_content.loc[arxiv_paper_content["categories"] == "physics.gen-ph",:]
astro_ph_papers
#doing .describe() shows 60303 different categories, the top being astro-ph for astrophysics
```
